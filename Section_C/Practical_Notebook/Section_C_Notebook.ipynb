```python
import pandas as pd

# =====================================================================
# TASK 1 & 2: Load Dataset and Data Audit (With Robust Delimiter & Space Fix)
# =====================================================================
print("--- TASK 1 & 2: DATA LOADING & AUDIT ---")

# sep=None automatically detects tabs (\t) or commas (,)
df = pd.read_csv('week_2_3_exam_student_signals.csv', sep=None, engine='python')

# Strip any hidden whitespaces or tab characters from the column names
df.columns = df.columns.str.strip()

# Print metrics to verify it successfully loaded all 8 columns
print(f"Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print("\nVerified Column Names:", list(df.columns))
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values Count:")
print(df.isnull().sum())
print("\nFirst Rows of Processed Data:")
print(df.head())

# =====================================================================
# TASK 3: Readiness Score Engine Function
# =====================================================================
print("\n" + "="*50 + "\n--- TASK 3: APPLYING READINESS SCORE ENGINE ---")
def calculate_readiness(row):
    # Dynamic Weightage: Attendance (20%), Tasks (30%), Assignment (25%), Lab (25%)
    score = (row['attendance_pct'] * 0.20) + \
            (row['tasks_completed'] * 3.0) + \
            (row['assignment_score'] * 0.25) + \
            (row['lab_score'] * 0.25)
    return min(score, 100) # Capped at 100%

# Generate final_score calculation
df['final_score'] = df.apply(calculate_readiness, axis=1)
print(df[['name', 'final_score']])

# =====================================================================
# TASK 4: Classification Logic
# =====================================================================
print("\n" + "="*50 + "\n--- TASK 4: CLASSIFYING PERFORMANCE BANDS ---")
def classify_band(score):
    if score >= 90: return 'Excellent'
    elif score >= 70: return 'Strong'
    elif score >= 40: return 'Developing'
    else: return 'Support'

# Generate performance bands
df['band'] = df['final_score'].apply(classify_band)
print(df[['name', 'final_score', 'band']])

# =====================================================================
# TASK 5: Analytics Outputs
# =====================================================================
print("\n" + "="*50 + "\n--- TASK 5: GENERATING ANALYTICS INSIGHTS ---")

# 1. Department Performance Metric
print("1. Department Average Scores:")
print(df.groupby('department')['final_score'].mean())

# 2. Top Performer Identification
top_student = df.loc[df['final_score'].idxmax()]
print(f"\n2. Top Performer: {top_student['name']} ({top_student['final_score']:.2f}%)")

# 3. Targeted Support/Intervention List
support_list = df[df['band'] == 'Support']
print("\n3. Students Needing Support:")
if not support_list.empty:
    print(support_list[['name', 'department', 'final_score']])
else:
    print("None! All students scored above the support threshold.")
```
